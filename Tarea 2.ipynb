*Ejercicio 1*

Dependiendo del número de materias que lleva un alumno  y del número de créditos que tenga cada materia, se hará un cálculo de la inscripción:

 - Si lleva menos de 8 créditos no pagará inscripción
 - Si  la suma de créditos es igual o mayor a 8 y menor a 20 créditos deberá pagar 1,000.00 de inscripción. 
 - Si la suma es de 20 créditos o más deberá pagar 2,000.00 de inscripción.
 
Pedir al alumno el número de créditos y desplegar lo que debe pagar de inscripción

```python
# Programa

print("Asignaturas para cursar durante el semestre:")
print("quimica = 5")
print("matematicas = 3")
print("fisica = 6")
print("programacion = 9")
print("filosofia = 2")
print("historia = 4")
print("estadistica = 6")
print("ingles = 1")

m1 = input("Ingrese una materia: ")
m2 = input("Ingrese la siguiente materia: ")
m3 = input("Ingrese la siguiente materia: ")

while m1 == m2 or m1 == m3 or m2 == m3:
    print("No puedes ingresar asignaturas repetidas, revisa de nuevo")
    m1 = input("Ingrese una materia: ")
    m2 = input("Ingrese la siguiente materia: ")
    m3 = input("Ingrese la siguiente materia: ")

if m1 == "quimica":
    m1 = 5
elif m1 == "matematicas":
    m1 = 3
elif m1 == "fisica":
    m1 = 6
elif m1 == "programacion":
    m1 = 9
elif m1 == "filosofia":
    m1 = 2
elif m1 == "historia":
    m1 = 4
elif m1 == "estadistica":
    m1 = 6
elif m1 == "ingles":
    m1 = 1

if m2 == "quimica":
    m2 = 5
elif m2 == "matematicas":
    m2 = 3
elif m2 == "fisica":
    m2 = 6
elif m2 == "programacion":
    m2 = 9
elif m2 == "filosofia":
    m2 = 2
elif m2 == "historia":
    m2 = 4
elif m2 == "estadistica":
    m2 = 6
elif m2 == "ingles":
    m2 = 1

if m3 == "quimica":
    m3 = 5
elif m3 == "matematicas":
    m3 = 3
elif m3 == "fisica":
    m3 = 6
elif m3 == "programacion":
    m3 = 9
elif m3 == "filosofia":
    m3 = 2
elif m3 == "historia":
    m3 = 4
elif m3 == "estadistica":
    m3 = 6
elif m3 == "ingles":
    m3 = 1

creditos = m1 + m2 + m3

if creditos < 8:
    print ("No pagas inscripción por el momento")
elif creditos < 20:
    print ("Deberás pagar la cantidad de $1,000.00")
else:
    print ("Deberás pagar la cantidad de $2,000.00")
```
